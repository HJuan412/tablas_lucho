```python
# pandas es el modulo que se usa en python para trabajar con dataframes
# numpy es el modulo para trabajar con matrices
import pandas as pd
import numpy as np
# cargamos la tabla de datos
tab = pd.read_csv('/home/hernan/Documents/example_tab.txt', sep='\t')
tab
```

```python
# extraemos los genes Ãºnicos
genes = np.unique(tab)
genes
```

```python
# creamos la nueva tabla, usando los genes como indice, los experimentos como nombres de columnas
# todos los valores son 0 por defecto
tab2 = pd.DataFrame(0, index=genes, columns=tab.columns)
tab2
```

```python
# asignamos el valor 1 a los genes observados en cada experimento
for exp in tab.columns:
    tab2.loc[tab[exp], exp] = 1
tab2
```
